```python
import nltk
from nltk.collocations import *
from ir import ReceiptIter
from ir import make_ngrams
```

```python
bigram_measures = nltk.collocations.BigramAssocMeasures()
```

```python
trigram_measures = nltk.collocations.TrigramAssocMeasures()
```

```python
sents = [sent for sent in ReceiptIter()]
```

```python
web_sents = sents[::2]
```

```python
words = [item for sublist in web_sents for item in sublist]
```

```python
len(words)
```

```python
finder = BigramCollocationFinder.from_words(words)
```

```python
finder.nbest(bigram_measures.pmi, 10)
```

```python
web_sents[:20]
```

```python
finder.nbest(bigram_measures.pmi, 200)
```

```python
tfinder = TrigramCollocationFinder.from_words(words)
```

```python
tfinder.nbest(bigram_measures.pmi, 10)
```

```python
tfinder.apply_freq_filter(3)
```

```python
tfinder.nbest(trigram_measures.pmi, 200)
```

```python
scored = finder.score_ngrams(bigram_measures.raw_freq)
```

```python
sorted([(score, bigram) for bigram, score in scored], reverse=True)
```

```python
scored_pmi = finder.score_ngrams(bigram_measures.pmi)
```

## This is the good result
